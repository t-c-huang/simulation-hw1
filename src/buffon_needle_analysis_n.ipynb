```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
data_dir = './buffon_needle_data_n'
```

```python
df = pd.DataFrame(columns=['number_of_drops', 'pi'])
for file in os.listdir(data_dir):
    df_read = pd.read_csv(os.path.join(data_dir, file))
    df = pd.concat([df, df_read[['number_of_drops', 'pi']]], ignore_index=True)
```

```python
df.groupby('number_of_drops').agg(['mean', 'std'])
```

```python
df
```

```python
df.boxplot(column='pi', by='number_of_drops', figsize=(10, 5))
```

```python
#plot histogram of different number of drops in the subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
for i, ax in enumerate(axes.flatten()):
    drop_num = df['number_of_drops'].unique()[i]
    ax.hist(df[df['number_of_drops'] == drop_num]['pi'], bins=25)
    ax.set_title('number of drops = {}'.format(drop_num))
    ax.set_xlabel('pi')
    ax.set_ylabel('frequency')
```
